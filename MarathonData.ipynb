```python
#rename columns
```

```python
#Year of event                  int64
#Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers      int64
#Athlete performance           object
#Athlete gender                object
#Athlete average speed        float64
#Athlete ID                     int64
#athlete_age                    int32
```

```python
df2 = df2.rename(columns = {
    'Year of event': 'year',
    'Event dates': 'race_day',
    'Event name': 'race_name',
    'Event distance/length': 'race_length',
    'Event number of finishers': 'race_number_of_finishers',
    'Athlete performance': 'athlete_performance',
    'Athlete gender': 'athlete_gender',
    'Athlete average speed': 'athlete_avg_speed',
    'Athlete ID': 'athlete_id'
})
```

```python
df2.head(10)
```

```python
#reorder columns
```

```python
df3 = df2[['race_day', 'race_name', 'race_length', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_avg_speed']]
```

```python
data = df3
```

```python
sns.lmplot(data, x = 'athlete_age', y = 'athlete_avg_speed', hue = 'athlete_gender')
```

```python
#Difference in speed for the 50k and 50mi male to female
```

```python
data.groupby(['race_length', 'athlete_gender'])['athlete_avg_speed'].mean()
```

```python
#What age groups are the best in the 50m Race (20 + races min)
```

```python
data.query('race_length == "50mi"').groupby('athlete_age')['athlete_avg_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count>19')
```

```python
#seasons for the data -> Slower in summer than winter?
# spring 3-5
# summer 6-8
# fall 9-11
# winter 12-2
```

```python
data['race_month'] = data['race_day'].str.split('.').str.get(1).astype(int)
```

```python
data['race_season'] = data['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```
